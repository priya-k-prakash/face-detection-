```python
img = cv2.cvtColor(img_raw,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap = 'gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('/home/priya/Downloads/frontalFace10/haarcascade_frontalface_default.xml')
```

```python
faces_rect = haar_cascade_face.detectMultiScale(img_gray,scaleFactor = 1.2,minNeighbors = 5)
```

```python
for (x,y,w,h) in faces_rect:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(img)
```
